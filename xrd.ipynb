```python
from pymatgen.ext.matproj import MPRester
from pprint import pprint
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("MSue3C8DVEAaHs0r") as m:
    struc = m.get_structure_by_material_id("mp-9362")
    xrd=XRDCalculator(1.54060)
    patt=xrd.get_pattern(struc,two_theta_range=(0, 30))
    
    hkl=patt.hkls
    dhkl=patt.d_hkls
    
    theta2=patt.x
    inten=patt.y
```

Premier pic

```python
print('hkl:',hkl[0])
print('distance:', dhkl[0])
print('angle:', theta2[0])
print('intensité:',inten[0])
```

Deuxième pic

```python
print('hkl:',hkl[1])
print('distance:', dhkl[1])
print('angle:', theta2[1])
print('intensité:',inten[1])
```

Comme ce pic est identique que le précédent, nous allons plutôt nous intéresser au suivant.

```python
print('hkl:',hkl[2])
print('distance:', dhkl[2])
print('angle:', theta2[2])
print('intensité:',inten[2])
```
