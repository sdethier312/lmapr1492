```python
from pymatgen.ext.matproj import MPRester
import numpy as np
from pprint import pprint
from pymatgen.electronic_structure.plotter import BSPlotter as bsp
with MPRester("MSue3C8DVEAaHs0r") as m:
    band = m.get_bandstructure_by_material_id("mp-9362")
    gap=band.get_band_gap()
    print('bande interdite:',gap)
    print('Energie de Fermi:',band.efermi)

    bs=bsp(band)
    bs.get_plot()
    d=bs.bs_plot_data()
    
    vb=d["vbm"]
    cb=d["cbm"]
    
    en=d["energy"]
    tick=d["ticks"]

    print('bande de valence:',vb)
    print('bande de conduction:',cb)
    print('distances=',tick["distance"])
  
```
